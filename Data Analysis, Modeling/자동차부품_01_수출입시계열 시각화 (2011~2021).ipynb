```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 전처리한 데이터 불러오기
df = pd.read_csv('../Raw 데이터/자동차부품(742)_수출입통계_시계열용 (2011~2021).csv')

# 사용할 데이터만 남기기 (시점, 수출액, 수입액)
df = df.loc[1:]
df.rename(columns=df.iloc[0], inplace=True)
df = df.drop(df.index[0])
df = df[['년월', '수출 금액', '수입 금액']] # 필요한 열만 사용

df.columns = ['시점', '수출액', '수입액']
df.reset_index(drop=True, inplace=True)
df
```

```python
# 날짜데이터 변환 (월별 데이터이므로 가장 마지막날로 설정)
df['시점'] = pd.date_range('2011-01-31', periods=132, freq='M')
df
```

```python
# 그래프 내 한글 깨짐

import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':  # 맥OS 
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # 윈도우
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...  sorry~~~')
```

```python
# 수출액, 수입액: 수치형으로 변환
df['수출액']=df['수출액'].astype(float)
df['수입액']=df['수입액'].astype(float)
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='시점', y='수출액', label='수출액')
sns.lineplot(data=df, x='시점', y='수입액', color='r', label='수입액')
plt.xlabel('년도', fontsize=13)
plt.ylabel('금액', fontsize=13)
plt.title('자동차부품 년도별 수출입액 변화 추이 (단위: 천 달러)', fontsize=15)
plt.legend()

# 코로나 19 시점 표시
import datetime
x = [datetime.date(2020, 1, 31), datetime.date(2020, 1, 31)]
y = [0, 2.6e6]

plt.plot(x, y, color='green') 
```

### 2. 자동차부품 산업 수출액 변화 시각화

```python
# (2) 수출액 시각화
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='시점', y='수출액', label='수출액')
plt.xlabel('년도', fontsize=13)
plt.ylabel('금액', fontsize=13)
plt.title('자동차부품 년도별 수출액 변화 추이 (단위: 천 달러)', fontsize=15)

# 코로나 19 시점 표시
import datetime
x = [datetime.date(2020, 1, 31), datetime.date(2020, 1, 31)]
y = [0, 2.6e6]

plt.plot(x, y, color='green') 
```

- `2020년 코로나19 이후 자동차부품 수출액 하락` ; 2020년 상반기, 코로나19로 인한 부품수급 차질로 인해, 국내 완성차 생산공장 가동 중단에 이어 글로벌 완성차 판매 감소로 부품 수출이 급격하게 감소했다. 이외에도 전기차, 수소차로의 변환 등에 따라 생산 부품의 구조적인 수요 감소가 일어나고 있다.


- `2021년 넘어가는 시점부터 수출액 회복` : 코로나 19로 인한 생산가동중단, 차량용 반도체 수급난에도 불구하고, 코로나19로 인한 기저효과와 전세계 자동차 판매 호조 및 국내 브랜드 점유율 상승에 따라 수출액을 회복하고 있다.

### 3. 자동차부품 산업 수입액 변화 시각화

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='시점', y='수입액', color='r', label='수입액')
plt.xlabel('년도', fontsize=13)
plt.ylabel('금액', fontsize=13)
plt.title('자동차부품 년도별 수입액 변화 추이 (단위: 천 달러)', fontsize=15)

# 코로나 19 시점 표시
import datetime
x = [datetime.date(2020, 1, 31), datetime.date(2020, 1, 31)]
y = [0, 600000]

plt.plot(x, y, color='green') 
```

- `2011~2012년 자동차부품 수입액이 높은 이유` : 한미 FTA와 한EU FTA가 진행되는 시기이므로, 관세가 줄어들고 수입수출이 활성화되었다.


- `2020년 하반기부터 수입액이 꾸준히 증가한 이유` : 우리나라에 있는 자동차부품 산업의 대부분은 내연기관 위주의 업체이다. 하지만, 내연기관 기술만큼 친환경차 기술은 아직까지 큰 격차를 보이고 있는데, 코로나 19가 시작된 전후로 전기차, 수소차와 같은 친환경차가 각광받으면서 해당 자동차에 필요한 자동차 부품을 수입하기 시작했다. 따라서 이에 대한 부품, 소재의 국산화 필요성을 뉴스에서 다수 제시하고 있는 상황이다.
