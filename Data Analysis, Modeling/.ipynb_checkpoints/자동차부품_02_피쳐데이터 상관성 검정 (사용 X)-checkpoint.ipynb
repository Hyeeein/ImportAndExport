```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
# 데이터 불러오기
df = pd.read_csv('../Data Preprocessing/자동차부품 데이터셋.csv', encoding='cp949')
df = df.iloc[:, 1:]
df
```

```python
df.info()
```

```python
# 수출액, 수입액, 환율은 float형으로  수치형으로 변환
df['수출액']=df['수출액'].astype(float)
df['수입액']=df['수입액'].astype(float)
df['환율']=df['환율'].astype(float)
```

```python
df.info()
```

```python
# 둘 사이에 상관성이 있으면 안됨 (유의확률이 0.01보다는 커야함)
col = df.columns

for i in range(3, len(col)):
    for j in range(3, len(col)):
        corr = stats.pearsonr(df[col[i]], df[col[j]])
        print(f'{col[i]}과 {col[j]} : {corr}')
        
# 환율과 국제유가
# 소비자 물가지수와 가동률지수(원지수)
# 국제유가와 고용률(%)

# 둘다 겹치는 국제유가는 빠이...
# 수출입 경쟁력 예측에 더 유리한 변수를 놔둠 : 소비자 물가지수 탈락... 안녕...
```
