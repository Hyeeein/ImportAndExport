```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
employment_ratio = pd.read_csv('datasets/고용률 (2015~2021).csv', encoding='cp949')
employment_ratio = employment_ratio.drop(employment_ratio.index[0])
employment_ratio.reset_index(drop=True, inplace=True)
employment_ratio
emp = employment_ratio.loc[0:83, ['계']]
emp
```

```python
df7 = pd.read_csv('datasets/고용률 (2015~2021).csv', encoding='cp949')
df7 = df7.loc[1:84, ['계']]
df7.columns = ['고용률(%)']
df7.reset_index(drop=True, inplace=True)
df7
```

```python
oil = pd.read_csv('datasets/국제원유가격(2015~2021).csv', encoding='utf-8')
oil
dubai_oil = oil['Dubai']
dubai_oil
```

```python
aircost = pd.read_csv('datasets/반도체 항공운임(2015_2021).csv', encoding='utf-8')
# aircost['mean'] = aircost.mean(axis='columns')
aircost.sort_values(by='년월', inplace=True)
aircost['mean'] = aircost.iloc[:,1:].mean(axis=1)
aircost.reset_index(drop=True, inplace=True)
aircost[['년월', 'mean']]
aircost_mean = aircost['mean']
aircost_mean
```

```python
import_export = pd.read_csv('datasets/반도체(731)_수출입통계(2015~2021).csv', encoding='utf-8')
import_export = import_export.loc[1:]
import_export.rename(columns=import_export.iloc[0], inplace=True)
import_export = import_export.drop(import_export.index[0])
import_export = import_export[['년월', '수출 금액', '수입 금액']]
import_export.reset_index(drop=True, inplace=True)
import_export
```

```python
utilization_rate = pd.read_csv('datasets/반도체_가동률지수(2011~2021).csv', encoding='utf-8')
utilization_rate = utilization_rate.iloc[48:,:]
utilization_rate.reset_index(drop=True, inplace=True)
util = utilization_rate['가동률지수(원지수)']
```

```python
CPI = pd.read_csv('datasets/소비자 물가지수 (2015~2021).csv', encoding='cp949')
cpi = CPI['전국']
cpi
```

```python
interest_rate = pd.read_csv('datasets/한국은행_기준금리(2015.01~2021.12).csv', encoding='cp949')
interest_rate = interest_rate.iloc[:84,:2]
interest = interest_rate['금리(연%)']
interest
```

```python
exchange_rate = pd.read_csv('datasets/환율 (2015_2021).csv', encoding='utf-8')
exchange_rate = exchange_rate.loc[1:]
exchange_rate = exchange_rate.iloc[:,:2]
exchange_rate.reset_index(drop=True, inplace=True)
won_dollar = exchange_rate['원/달러']
won_dollar
```

```python
# 데이터 셋 만들기
semiconductor = pd.concat([import_export, won_dollar, cpi, interest, dubai_oil, aircost_mean, emp, util], axis=1)
```

```python
semiconductor
```

```python
semiconductor = semiconductor.rename(columns={'년월':'시점', '수출 금액':'수출액', '수입 금액':'수입액', '원/달러':'환율', '전국':'소비자 물가지수', '금리(연%)':'기준금리(연%)', 'Dubai':'국제유가', 'mean':'항공 운임', '계':'고용률(%)', '가동률지수(원지수)':'가동률지수(원지수)'})
```

```python
semiconductor.to_csv('반도체 데이터셋.csv', encoding='cp949')
```
